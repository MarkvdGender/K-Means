```python
r = requests.get("http://localhost:8080/user/all")
```

```python
r.json()
```

```python
def get_mean(user):
    temp_df = pd.DataFrame(user['trips'])
    user_id = user['id']
    mean_afstand = temp_df['MeasureDistanceMetres'].mean()
    mean_uitstoot = temp_df['MeasureCo2Grams'].mean()
    user_row = pd.DataFrame([[user_id, mean_afstand, mean_uitstoot]], columns=['user', 'afstand', 'uitstoot'])
    return user_row
```

```python
test_user = r.json()[0]
```

```python
get_mean(test_user)
```

```python
for user in r.json():
    user_df = user_df.append(get_mean(user), ignore_index=True)
```
