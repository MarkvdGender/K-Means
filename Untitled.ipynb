# Inladen van de Trips

```python
r = requests.get("http://localhost:8080/trip/all")
```

```python
trips = r.json()
```

```python
def fuse_trips(trip1, trip2):
    measure_distance_metres = trip1['MeasureDistanceMetres']+trip2['MeasureDistanceMetres']
    measure_co2_grams = trip1['MeasureCo2Grams']+trip2['MeasureCo2Grams']
    measure_minutes_exercise = trip1['MeasureMinutesExercise']+trip2['MeasureMinutesExercise']
    measure_co2_grams_km = measure_co2_grams/(measure_distance_metres/1000)

    trip_dict = {
        "MeasureDistanceMetres": measure_distance_metres,
        "MeasureCo2Grams": measure_co2_grams,
        "MeasureMinutesExercise": measure_minutes_exercise,
        "MeasureCo2GramsKm": measure_co2_grams_km,
        "EventStartDateTimeUtc" : start.strftime('%Y-%m-%d %H:%M:%S'),
        "EventEndDateTimeUtc": end.strftime('%Y-%m-%d %H:%M:%S')}
    
    return trip_dict
```

```python
def trip_merge(lijst):
    global start
    global end

    for i in range (len(lijst)):
        start = datetime.strptime(lijst[i]['EventStartDateTimeUtc'], '%Y-%m-%d %H:%M:%S')
        if(abs(end-start) <= timedelta(minutes=5) and i != 0):
            trip1, trip2 = lijst[i-1], lijst[i]
            trip = fuse_trips(trip1, trip2)
            lijst.pop(i), lijst.pop(i-1)
            lijst.append(trip)
            end = datetime.strptime(lijst[i]['EventEndDateTimeUtc'], '%Y-%m-%d %H:%M:%S')
            return trip_merge(lijst)
        else:
            end = datetime.strptime(lijst[i]['EventEndDateTimeUtc'], '%Y-%m-%d %H:%M:%S')
            
# end of recursion
    return lijst
```

```python
trips = trip_merge(trips)
```

```python
trip_df = pd.DataFrame()
for trip in trips:
    trip_df = trip_df.append(trip, ignore_index=True)
```

# K-means

```python
plt.title('Histogram')
plt.xlabel('MeasureDistanceMetres')
plt.ylabel('Trips')
plt.hist(trip_df['MeasureDistanceMetres'], 100)
plt.show()
```

```python
def train_kmeans(clusters, df):
    return KMeans(n_clusters = clusters, random_state=0).fit(df)
```

```python
kmeans_df = trip_df.drop(['MaskedUserId', 'EventEndDateTimeLocal', 'EventEndDateTimeUtc', 'EventId', 'EventModality',
                        'EventOffPeak', 'EventPurpose', 'EventStartDateTimeLocal', 'EventStartDateTimeUtc', 'EventType',
                          'MeasureAverageKMH', 'MeasureCo2GramsKm', 'MeasureCo2Grams', 'MeasureMinutesExercise'], axis=1)
```

```python
inertias=[]

for i in range(1,10):
    kmodel = train_kmeans(i, kmeans_df)
    inertias.append(kmodel.inertia_)
```

```python
plt.plot(range(1,10), inertias, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show() 
```

```python
kmodel = train_kmeans(4, kmeans_df)
```

```python
trip_df['groep'] = kmodel.labels_
```

```python
def plot(x, y, highlight):
    plt.scatter(x, y,c=highlight, cmap='plasma')
    plt.xlim(0, max(x))
    plt.ylim(0, max(y))

    plt.show()
```

```python
plot(trip_df['MeasureDistanceMetres'], trip_df['MeasureCo2Grams'], trip_df['groep'])
```
