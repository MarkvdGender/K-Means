```python
import pandas as pandas
```

```python
sensing = pandas.read_json("sensing-sample-2-trips.json")
```

```python
sensing.head(10)
```

```python
for key, value in sensing['reading'][0].items():
    i=0
    values=[]
    for row in sensing.iterrows():
        values.append(sensing.reading[i].get(key))
        i+=1
    sensing[key] = values
```

```python
amount=[]
emissionType=[]
i=-1
for row in sensing.iterrows():
    i+=1
    try:
        amount.append(sensing['totalEmission'][i][0].get('amount'))
        emissionType.append(sensing['totalEmission'][i][0].get('emissionType'))
    except TypeError:
        amount.append(None)
        emissionType.append(None)
        continue
        
sensing['emissionType'] = emissionType
sensing['amount'] = amount
```

```python
del sensing['reading']
```

```python
del sensing['totalEmission']
```
